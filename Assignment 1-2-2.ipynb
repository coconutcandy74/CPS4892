```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_and_show_histogram(image, title):
    histogram = np.zeros(256, dtype=int)
    for row in image:
        for pixel in row:
            histogram[pixel] += 1
    
    plt.figure(figsize=(10, 5))
    plt.title(title)
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.bar(range(256), histogram, width=1.0, edgecolor='black')
    plt.show()
```

```python
def convert_to_grayscale(image_path):

    image = cv2.imread(image_path)
    
    image_array = np.array(image)
    
    height, width, channels = image_array.shape
    
    gray_image = 0.299 * image_array[:, :, 2] + 0.587 * image_array[:, :, 1] + 0.114 * image_array[:, :, 0]
    gray_image = gray_image.astype(image_array.dtype)
    
    return gray_image
```

```python
def histogram_equalization(image_path):

    gray_image = convert_to_grayscale(image_path)
    
    calculate_and_show_histogram(gray_image, "Original Histogram")
    
    equalized_image = cv2.equalizeHist(gray_image)
    
    calculate_and_show_histogram(equalized_image, "Equalized Histogram")
    
    plt.figure(figsize=(10, 5))
    plt.subplot(1, 2, 1)
    plt.title("Original Image")
    plt.imshow(gray_image, cmap='gray')
    plt.axis('off')

    plt.subplot(1, 2, 2)
    plt.title("Equalized Image")
    plt.imshow(equalized_image, cmap='gray')
    plt.axis('off')

    plt.show()

image_path = r'C:\Users\Coconut\Desktop\CPS4892\Assignment-1-2-HE-source.jpeg'
histogram_equalization(image_path)
```
