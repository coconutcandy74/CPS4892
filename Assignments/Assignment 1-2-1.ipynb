```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def convert_to_grayscale_and_show_histogram(image_path):

    image = cv2.imread(image_path)
    image_array = np.array(image)
    height, width, channels = image_array.shape
    
    gray_image = 0.299 * image_array[:, :, 2] + 0.587 * image_array[:, :, 1] + 0.114 * image_array[:, :, 0]
    gray_image = gray_image.astype(image_array.dtype)
    
    histogram = np.zeros(256, dtype=int)
    for row in gray_image:
        for pixel in row:
            histogram[pixel] += 1
    
    plt.figure(figsize=(10, 5))
    plt.title("Histogram")
    plt.xlabel("Pixel Value")
    plt.ylabel("Frequency")
    plt.bar(range(256), histogram, width=1.0, edgecolor='black')
    plt.show()

image_path = r'C:\Users\Coconut\Desktop\CPS4892\Assignment-1-2-HE-source.jpeg'
convert_to_grayscale_and_show_histogram(image_path)
```
